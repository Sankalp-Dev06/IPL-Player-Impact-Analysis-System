```python
import json

file_path = 'D:\\Code\\Context-Aware IPL Player Impact Prediction System\\data\\ipl_json\\1136606.json'

with open (file_path,'r') as f:
    data = json.load(f)

```

```python
season = data["info"]["season"]
match_date = data["info"]["dates"][0]
venue = data["info"]["venue"]
teams = data["info"]["teams"]
toss_winner = data["info"]["toss"]["winner"]
toss_decision = data["info"]["toss"]["decision"]

winner = data["info"].get("outcome", {}).get("winner")

print("Season:", season)
print("Date:", match_date)
print("Venue:", venue)
print("Teams:", teams)
print("Toss Winner:", toss_winner)
print("Toss Decision:", toss_decision)
print("Winner:", winner)
type(season)
```

```python
first_inning = data["innings"][0]
first_over = first_inning["overs"][0]
first_delivery = first_over["deliveries"][0]

print(first_delivery)
```
